# Cluster Accuracy

```python
test_index = [i for i in range(1000) if (i % 5 == 0)]

train_index = [i for i in range(1000) if i not in test_index]
```

```python
test_data = train_feature.loc[test_index]

train_data = train_feature.loc[train_index]
```

```python
from collections import Counter
d_10 = {}

for index, row in test_data.iterrows():
    d_10[row["cluster"]] = {}
    d_10[row["cluster"]]["test_group"] = row["PCA_500_cluster_10_without_concat"]
    d_10[row["cluster"]]["train_group"] = []

for index,row in train_data.iterrows():
    d_10[row["cluster"]]["train_group"].append(row["PCA_500_cluster_10_without_concat"])
    c = Counter(d_10[row["cluster"]]["train_group"])
    value, count = c.most_common()[0]
    d_10[row["cluster"]]["major"] = value

i = 0
for key in d_10.keys():
    if d_10[key]["test_group"] == d_10[key]["major"]:
        i+= 1

print(i / len( d_10.keys()))
    
```

```python
def cluster_acc(col_name):
    d = {}

    for index, row in test_data.iterrows():
        d[row["cluster"]] = {}
        d[row["cluster"]]["test_group"] = row[col_name]
        d[row["cluster"]]["train_group"] = []

    for index,row in train_data.iterrows():
        d[row["cluster"]]["train_group"].append(row[col_name])
        c = Counter(d[row["cluster"]]["train_group"])
        value, count = c.most_common()[0]
        d[row["cluster"]]["major"] = value

    i = 0
    for key in d.keys():
        if d[key]["test_group"] == d[key]["major"]:
            i += 1



    print("Cluster acc:",i / len( d.keys()))

    #return d
    
```

```python
cluster_acc("PCA_500_cluster_10_without_concat")
cluster_acc("PCA_500_cluster_20_without_concat")
cluster_acc("PCA_500_cluster_2_without_concat")
```

```python
cluster_acc("PCA_100_cluster_20_without_concat")
cluster_acc("PCA_100_cluster_2_without_concat")
```
